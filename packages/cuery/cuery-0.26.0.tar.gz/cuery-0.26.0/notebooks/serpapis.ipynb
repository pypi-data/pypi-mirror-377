# HasData

```python
import asyncio
import importlib

import aiohttp
from tqdm.auto import tqdm

from cuery.search import SearchResult
from cuery.seo.aio import hasdata
from cuery.utils import set_env

set_env(return_vars=False)

importlib.reload(hasdata)
```

```python
aio = await hasdata.query("What is the ISS?", validate=True, country="us")
print(aio)
```

```python
async def progress_callback(status: dict):
    print("Progress update:", status["n"], "/", status["total"])


prompts = [
    "what is intermittent fasting benefits",
    "difference between IPv4 and IPv6",
    "how to clean a air fryer safely",
    "tips for improving sleep hygiene",
    "understanding compound interest formula",
    "can dogs eat grapes toxic effects",
    "best practices for remote team communication",
    "how SEO affects website traffic",
    "importance of data privacy laws",
    "what causes climate change explained",
]

pbar = tqdm(total=len(prompts))

policies = {
    "timeout": 10,
    "n_concurrent": 10,
    "retries": 3,
    "wait_max": 10,
    "fallback": SearchResult.fallback(),
    "pbar": pbar,
    "progress_callback": progress_callback,
    "min_iters": 2,
}

async with aiohttp.ClientSession() as session:
    coros = await hasdata.gather(prompts, session=session, policies=policies, execute=False)
    results = await asyncio.gather(*coros)
```

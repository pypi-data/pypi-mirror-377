### Parse chapter by blocks


```python
page = doc[6]
```

```python
def is_list_item(text):
    return text.strip().startswith(('-', '*', '•', '1.', 'a.', 'i.'))
```

```python
blocks = page.get_text("dict")["blocks"]
for block in blocks:
    for line in block["lines"]:
        for span in line["spans"]:
            text = span["text"]
            font_flags = span["flags"]
            font_size = span["size"]
            # 判断段落类型
            if "bold" in span["font"]:  # 检查加粗
                print("Bold text:", text)
            elif font_size > 10:  # 检查标题
                print("Title:", text)
            elif "underline" in span["font"]:
                print("Underline text:", text)
            elif is_list_item(text):
                print("List item:", text)
            else:
                print("Regular text:", text)
```

```python
blocks = page.get_text("dict")["blocks"]
for i, block in enumerate(blocks):
    print(f'=== block [{i}] ===')
    for line in block["lines"]:
        for span in line["spans"]:
            text = span["text"]
            font_flags = span["flags"]
            font_size = span["size"]
            # 判断段落类型
            if "bold" in span["font"]:  # 检查加粗
                print("Bold text:", text)
            elif font_size > 10:  # 检查标题
                print("Title:", text)
            elif "underline" in span["font"]:
                print("Underline text:", text)
            elif is_list_item(text):
                print("List item:", text)
            else:
                print("Regular text:", text)
```

```python
def print_block_lines(block):
    for line in block["lines"]:
        line_text = ''.join(span["text"] for span in line["spans"])
        print(line_text)
```

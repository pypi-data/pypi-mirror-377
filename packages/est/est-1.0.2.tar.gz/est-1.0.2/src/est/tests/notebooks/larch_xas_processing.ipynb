# XrayLarch XAS data processing

XrayLarch is a spectroscopy library for python: https://github.com/xraypy/xraylarch

## Example EXAFS spectrum

```python
%matplotlib inline
import matplotlib.pyplot as plt

from est.tests.data import example_spectrum

energy, mu = example_spectrum("exafs", "EXAFS_Cu.dat")

plt.figure(figsize=(20, 10))
plt.plot(energy, mu)
plt.xlabel("Energy (eV)")
plt.ylabel("mu")
print("Example EXFAS spectrum")
```

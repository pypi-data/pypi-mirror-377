# Loading some modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
```

# Checking blue and red centrals luminosity limits as a function of redshift

```python
blue_cen_lum_pdf=np.loadtxt('blue_cen_lum_pdf.txt')
```

```python
for i in range(0,len(blue_cen_lum_pdf.T[0])):
    plt.plot([blue_cen_lum_pdf.T[0,i],blue_cen_lum_pdf.T[0,i]],
             [np.log10(blue_cen_lum_pdf.T[1,i]),np.log10(blue_cen_lum_pdf.T[2,i])],color='blue')
plt.xlabel('redshift',fontsize=15)
plt.ylabel(r'log$_{10}(L)$',fontsize=15)
```

```python
red_cen_lum_pdf=np.loadtxt('red_cen_lum_pdf.txt')
```

```python
for i in range(0,len(red_cen_lum_pdf.T[0])):
    plt.plot([red_cen_lum_pdf.T[0,i],red_cen_lum_pdf.T[0,i]],
             [np.log10(red_cen_lum_pdf.T[1,i]),np.log10(red_cen_lum_pdf.T[2,i])],color='red')
plt.xlabel('redshift',fontsize=15)
plt.ylabel(r'log$_{10}(L)$',fontsize=15)
```

# Checking the fraction of red galaxies

```python
f_red=np.loadtxt('f_red.txt')
```

```python
plt.plot(f_red.T[0],f_red.T[1],color='red')
plt.xlabel('redshift',fontsize=15)
plt.ylabel(r'fraction of red galaxies',fontsize=15)
```

# Checking the blue satellite, red satellite, and red central luminosity pdfs as a function of redshift

```python
hdul=fits.open('bluesat_lum.fits')
```

```python
z=hdul[1].data['z']
loglum=hdul[1].data['loglum']
xedges=np.linspace(0,1.4,100)
yedges=np.linspace(7,11,100)
heatmap, xedges, yedges = np.histogram2d(z, loglum, bins=(xedges,yedges))
```

```python
plt.figure(figsize=(10,10))
plt.clf()
plt.imshow(heatmap.T,interpolation='nearest',origin='low',
           extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],aspect='auto')
plt.xlabel('redshift',fontsize=15)
plt.ylabel(r'log$_{10}(L)$',fontsize=15)
```

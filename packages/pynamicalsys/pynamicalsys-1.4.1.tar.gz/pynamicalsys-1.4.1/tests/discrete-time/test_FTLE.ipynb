```python
import numpy as np
import matplotlib.pyplot as plt
from pynamicalsys import DiscreteDynamicalSystem as dds
```

```python
ds = dds(model="standard map")
u = np.array([0.2, 0.5])
parameters = np.array([1.5])
N = int(1e6)
n = 100
```

```python
%%time
ftle = ds.finite_time_lyapunov(u, N, n, parameters=parameters)
```

```python
plt.hist(ftle[:, 0], bins="auto", density=True);
plt.hist(ftle[:, 1], bins="auto", density=True);
plt.xlim(-0.6, 0.6)
```

```python
%%time
ftle = ds.finite_time_lyapunov(u, N, n, parameters=parameters, num_exponents=1)
```

```python
plt.hist(ftle[:, 0], bins="auto", density=True);
plt.xlim(-0.6, 0.6)
```

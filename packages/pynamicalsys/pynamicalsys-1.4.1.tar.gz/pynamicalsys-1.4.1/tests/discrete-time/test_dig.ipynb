```python
import numpy as np
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
from pynamicalsys import DiscreteDynamicalSystem as dds
```

```python
k = 1.5
x = 0.5
ny = 1000
y = np.linspace(0, 1, ny)
total_time = int(1e4)
ds = dds(model="standard map")
```

```python
total_time, type(total_time)
```

```python
dig = Parallel(n_jobs=-1)(delayed(ds.dig)(np.array([x, y[i]]), total_time, parameters=k) for i in range(ny))
dig = np.array(dig)
```

```python
lypnv = Parallel(n_jobs=-1)(delayed(ds.lyapunov)(np.array([x, y[i]]), total_time, parameters=k) for i in range(ny))
lypnv = np.array(lypnv)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
ax[0].plot(y, lypnv[:, 0], "k")
ax[1].plot(y, dig, "k")
```

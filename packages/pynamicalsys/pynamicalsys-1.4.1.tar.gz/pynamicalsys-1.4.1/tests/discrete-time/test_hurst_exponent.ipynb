```python
import numpy as np
import matplotlib.pyplot as plt
from pynamicalsys import DiscreteDynamicalSystem as dds
```

```python
ds = dds(model="standard map")
```

```python
x = 0.5
y = 0.2
u = np.array([x, y])
k = np.linspace(2, 5, 1000)
total_time = 5000
```

```python
lyapunov = np.zeros((len(k), 2))
for i in range(len(k)):
    # Calculate Lyapunov exponent for each value of k
    lyapunov[i] = ds.lyapunov(u, total_time, parameters=k[i])
```

```python
H = np.zeros((len(k), 2))
for i in range(len(k)):
    # Calculate Lyapunov exponent for each value of k
    H[i] = ds.hurst_exponent(u, total_time, parameters=k[i])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 4), sharex=True)

ax[0].plot(k, lyapunov[:, 0], "k")

ax[1].plot(k, H[:, 0], "k")
ax[1].plot(k, H[:, 1], "r")
ax[1].plot(k, (H[:, 0] + H[:, 1]) / 2, "b")
```

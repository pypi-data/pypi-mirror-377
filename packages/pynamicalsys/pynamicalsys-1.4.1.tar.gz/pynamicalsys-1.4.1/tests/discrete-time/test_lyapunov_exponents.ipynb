```python
import numpy as np
import matplotlib.pyplot as plt
from pynamicalsys import DiscreteDynamicalSystem as dds
```

# Standard map

```python
u = np.array([0.2, 0.5])
k = 1.5
N = int(1e5)
sample_times = np.array(np.unique(
    np.logspace(np.log10(1), np.log10(N), 1000).astype(int)
), dtype=np.int64)
time = np.arange(1, N + 1)
np.random.seed(0)
```

```python
ds = dds(model="standard map")
```

```python
exponents = ds.lyapunov(u, 10, parameters=k, return_history=True)
```

```python
%%time
exponents = ds.lyapunov(u, N, parameters=k, return_history=True)
exponents2 = ds.lyapunov(u, N, parameters=k)
exponents[-1, :], exponents2
```

```python
%%time
exponents2 = ds.lyapunov(u, N, parameters=k, return_history=True, sample_times=sample_times)
```

```python
plt.plot(time, exponents[:, 0], lw=0.75)
plt.plot(sample_times, exponents2[:, 0], lw=0.75)
plt.xscale("log")
```

```python
plt.plot(time, exponents[:, 0], lw=0.75)
plt.plot(sample_times, exponents2[:, 0], lw=0.75)
plt.xscale("log")
```

# 4D symplectic map

```python
import numpy as np
import matplotlib.pyplot as plt
from pynamicalsys import DiscreteDynamicalSystem as dds

# Define the symplectic map
ds = dds(model="4d symplectic map")
```

```python
u = [3.0, 0, 0.5, 0]
eps1 = 0.5
eps2 = 0.1
xi = 0.001
parameters = [eps1, eps2, xi]

N = int(1e6)
sample_times = np.array(np.unique(
    np.logspace(np.log10(1), np.log10(N), 1000).astype(int)
), dtype=np.int64)
time = np.arange(1, N + 1)
np.random.seed(0)
lyapunov_exponents = ds.lyapunov(u, 10, parameters=parameters, return_history=True)
```

```python
%%time
lyapunov_exponents = ds.lyapunov(u, N, parameters=parameters, return_history=True)
```

```python
%%time
lyapunov_exponents2 = ds.lyapunov(u, N, parameters=parameters, return_history=True, sample_times=sample_times)
```

```python
plt.plot(time, lyapunov_exponents[:, 0], lw=0.75)
plt.plot(sample_times, lyapunov_exponents2[:, 0], lw=0.75)
plt.xscale("log")
```

```python
plt.plot(time, lyapunov_exponents[:, 0], lw=0.75)
plt.plot(sample_times, lyapunov_exponents2[:, 0], lw=0.75)
plt.xscale("log")
```

# Example of `seispack` usage with with Kepler red giants.
We use Kepler Light Curves Optimized For Asteroseismology (KEPSEISMIC) provided by Savita Mathur, Ângela Santos, and Rafael A. García and publically available on the MAST.

We will look at KIC 8684542.

```python
from seispack import *
kic=8684542
p=read_kepseismic_psd(kic)
```

If data are not locally available, it will download them.

Then, we can change the unit to microhertz and plot the spectrum

```python
p.to_xunit('muHz')
p.plot()
```

We include a quick estimate of the large separation Dnu through the autocorrelation techniques:

```python
dnu,numax,_,_=eacf(p)
```

We can use this value to plot an echelle diagram

```python
plot_echelle_image(p,dnu)
```
